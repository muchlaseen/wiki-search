```python
# import library
from bs4 import BeautifulSoup
import requests
import pandas as pd  
```

```python
# keyword
data_keyword = ['Artificial Intelligence','Machine Learning','Data Science','Information Retrieval']
```

```python
# url target
root_uri = 'https://en.wikipedia.org'
title = []
urls = []
word_timestamps = []
descriptions = []

for key in data_keyword:
    print('Item Keyword = ',key)
    print('==============================================================================')
    
    #url = root_uri+'/w/index.php?search='+key.replace(' ','+')+'&title=Special%3ASearch&profile=advanced&fulltext=1&ns0=1'
    url = root_uri+'/w/index.php?title=Special:Search&limit=1000&offset=0&ns0=1&search='+key.replace(' ','+')
    # response from url
    response = requests.get(url)
    # data abstract text
    data = response.text
    # parse html with soup
    soup = BeautifulSoup(data,'html.parser')
    # tags extraction from html parse
    content_list = soup.find('ul',{'class':'mw-search-results'})
    
    for item in content_list:
        heading_title = item.find('div',{'class':'mw-search-result-heading'})
        print('title = ',heading_title.find('a').get('title'))
        print('url = ',root_uri+heading_title.find('a').get('href'))
        word_timestamp = item.find('div',{'class':'mw-search-result-data'}).text
        print('word timestamp = ',word_timestamp)
        description = item.find('div',{'class':'searchresult'}).text
        print('description = ',description)
        title.append(heading_title.find('a').get('title'))
        urls.append(root_uri+heading_title.find('a').get('href'))
        word_timestamps.append(word_timestamp)
        descriptions.append(description)
    print('==============================================================================')
```

```python
# simpan data ke csv
dictFile = {
        'title':title,
        'description':descriptions,
        'word_timestamp':word_timestamps,
        'url':urls}
df = pd.DataFrame(dictFile) 
    
df.to_csv('data_wiki_articles.csv') 
```
