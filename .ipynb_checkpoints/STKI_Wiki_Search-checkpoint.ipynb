```python
from bs4 import BeautifulSoup
import urllib.request
import re
```

```python
root_uri  = "https://en.wikipedia.org/wiki/"
url = root_uri+"Artificial_intelligence"
```

```python
page = urllib.request.urlopen(url)
```

```python
try:
    page = urllib.request.urlopen(url)
except:
    print("Error, try again!.")
```

```python
# parsing halaman html dan print halaman
soup = BeautifulSoup(page, 'html.parser') 
print(soup)
```

```python
regex = re.compile('^tocsection-')
content_lis = soup.find_all('li', attrs={'class': regex})
print(content_lis)
```

```python
content = []
for li in content_lis:
    content.append(li.getText().split('\n')[0]) # ambil list dari konten wiki
print(content)
```

```python
with open('content.txt', 'w') as f:
    for i in content:
        f.write(i+"\n")
```

```python
see_also = []
for li in content_lis:
    a_tag = li.find('a', href=True) # find a tags that have a title and a class
    href = a_tag['href'] # ambil atribut href
    print('A = ',root_uri+href.replace('#',''))
    uri_real = root_uri+href.replace('#','')
    see_also.append(uri_real)
    
print(see_also)
```

```python
with open('see_also.csv', 'w') as f:
    for i in see_also:
        f.write(",".join(i)+"\n")
```
